create a scatter chart the no of bike share with temprature

```python
Figure, axes = plt.subplots(figsize =(140,60))

c = np.random.randint(1,10,len(bike))
s= np.random.randint(10,220, size =len(bike))


axes.scatter(bike["Count"],bike["Temprature"], s=s, c=c)

plt.xlabel("Count")

plt.show()
```

```python
Figure.savefig('Scatter.png')
```

Pie chart: Distribution of bike share with season

```python
Labels=["Spring","Summer","Autumn","Winter"]
Counts = Spring["Count"].sum(), Summer["Count"].sum(), Autumn["Count"].sum(), winter["Count"].sum()
```

```python
Figure2, Axes2 = plt.subplots(figsize =(14,8))

Axes2.pie(Counts, labels =Labels , autopct = "%1.1f%%" , explode =[0.1,0,0.1,0] )
Axes2.legend()
plt.show()
```
